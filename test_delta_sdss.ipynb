# Delta files downloaded from SDSS

```python
import astropy.io.fits as fits
import numpy as np
import astropy.constants as const
import matplotlib.pyplot as plt

delta_set = fits.open('delta-273.fits')

for i in range(1,11,1):
    d = delta_set[i].data
    lam = 10**d['LOGLAM']
    delta = d['DELTA']
    cont = d['CONT']

    plt.plot(lam,(delta+1)*cont)
    plt.plot(lam,cont)
    plt.xlabel(r'$\lambda/AA$')
    plt.ylabel(r'flux/ $10^{-17} ergs/s/cm^2/AA$')
    plt.title(i)
    plt.show()
```
