# unzip gz

```python
import gzip 
import os

filelist = os.listdir('./')
for file in filelist:
    if file.split('.')[-1]=='gz':
        f = file.replace('.gz','')
        g = gzip.GzipFile(file, "rb")  
        open(f, "wb+").write(g.read())
        g.close()
        os.remove(file)
```

# read fits

```python
import fitsio as fs
import matplotlib.pyplot as plt
import astropy.constants as const
import numpy as np

path = './delta_file/'
flist = os.listdir(path)

for f in flist:
    a = fs.FITS(path+f)
    num = len(a)-1
    for d in range(1,1+num):
        loglam = a[d]['LOGLAM'][:]
        delta = a[d]['DELTA'][:]
        cont = a[d]['CONT'][:]
        weight = a[d]['WEIGHT'][:]
        flux = (delta+1)*cont
        
        plt.plot(10**loglam,flux,label=r'$f(\lambda)=(\delta+1) C_q(\lambda) \bar{F}(z)$')
        plt.plot(10**loglam,cont,label = 'continuum')
        
        plt.title(f+':'+str(d))
        plt.legend()
        plt.show()
```
